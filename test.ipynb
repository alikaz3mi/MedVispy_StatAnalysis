```python
import simplebiostats as sb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_df = pd.read_csv(r'C:\Users\AliKazemi\Desktop\Features\Scalar_Features_metadata\ex2\Arcuate_Fasciculus.csv', index_col=0)
data_df.head()
```

```python
sb.summarize(data_df)
```

```python
for variable in data_df.columns:
    if variable != 'gender' and (data_df[variable].dtype != 'O'):
        sb.check_normality(data_df, variable)
```

```python
for variable in data_df.columns:
    if (data_df[variable].dtype != 'O'):
        sb.ttest_1samp(data_df, variable, hypothesis=0)
```

```python
sb.ttest_ind_by_group(data_df, 'laterality', 'curl')
```

```python
sb.ftest_std_by_group(data_df, 'laterality', 'curl')
```

```python
sb.bootstrap_by_group(data_df, group_var='laterality', variable='curl', n_iter=40)
```

```python
sb.ranksums_by_group(data_df, group_var='laterality', variable='curl')
```

```python
sb.plot_scatter_blandalt(input_df=data_df, var1='curl', var2='ad')
```

```python
sb.ttest_paired(data_df, var1='curl', var2='ad')
```

```python
sb.anova_by_group(data_df, resp_var='curl', group_var='laterality')
```

```python
sb.kruskal_by_group(data_df, resp_var='curl', group_var='laterality')
```

```python
sb.mcnemar_test(data_df, var1='laterality', var2='laterality')
```

```python
lr = sb.LinearRegression(data_df=data_df, resp_var='scan_age', cont_var='curl', cat_var='laterality', with_interactions=True)
```

```python
lr.fit()
```

```python
lr.check_model()
```

```python
lr.plot_predictions()
```
